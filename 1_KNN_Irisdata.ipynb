```python
from sklearn.datasets import load_iris
```

```python
iris_dataset = load_iris()
```

데이터셋 준비하기

```python
from sklearn.model_selection import train_test_split

# 훈련용 데이터, 테스트용 데이터, 훈련용 레이블, 테스트용 레이블
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'], test_size=0.3, random_state=0) # X는 대문자, y소문자로 쓰는 게 관행
```

```python
print('X_train 크기:', X_train.shape)
print('y_train 크기:', y_train.shape)

print('X_test 크기:', X_test.shape)
print('y_test 크기:', y_test.shape)
```

## 데이터 시각화

## 모델링

```python
# 지도학습--> Classification: k-최근접 알고리즘(KNN)
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
```

```python
# 훈련시키기(train)
knn.fit(X_train, y_train)

# -----> 훈련 완료되어 knn에 학습 결과가 들어감
```

```python
# 넘파이 배열 하나 만든 다음(임의의 수 샘플로), ---> 예측하기
import numpy as np
```

```python
# 최소한 2차원 데이터를 줘야 한다.
X_new = np.array([[5, 2.9, 1, 0.2]])
print(X_new)
```

```python
knn.predict(X_new)
```

```python
# 보기 좋게 만들기
prediction = knn.predict(X_new)
print('예측:', prediction)
print('예측한 타깃의 이름:', iris_dataset['target_names'][prediction])
```

```python
# 잘 만들었을까? 모델 평가하기
y_pred = knn.predict(X_test)
print('테스트 세트에 대한 예측값:\n', y_pred)
```

```python
# y_pred(예측값)과 y_test(실측값)을 비교한다.
y_pred == y_test

# False 하나 제외하고 모두 True가 나왔으므로 높은 정확도를 보임
```

```python
np.mean(y_pred == y_test) # 0.97이므로 97%의 정확도를 가진 모델이라고 말할 수 있다.
```

```python
print('테스트 세트에 대한 정확도:\n', np.mean(y_pred==y_test)*100, '%')
```

```python
# 평균치 말고 스코어로 정확도를 낼 수도 있다.
print('테스트 세트에 대한 정확도:\n', knn.score(X_test, y_test), '%')
```
